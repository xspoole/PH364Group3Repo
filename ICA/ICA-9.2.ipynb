## Part 1: Peaks of Interest

&#x2705; Based on your work in ICA 8.2, identify the peaks present in the `MuRun2010B.csv` data set.

```python
import pandas as pd

df = pd.read_csv("MuRun2010B.csv", skiprows=5)

df.head()
```

```python
# compute invariant mass
df['E_lab'] = df['E1'] + df['E2']
df['px_lab'] = df['px1 '] + df['px2']
df['py_lab'] = df['py1'] + df['py2']
df['pz_lab'] = df['pz1'] + df['pz2']

df['M_inv'] = (df['E_lab']**2 - (df['px_lab']**2 + df['py_lab']**2 + df['pz_lab']**2))**0.5

# display new columns
df.head()
```

```python
df['M_inv'].describe()
```

```python
import matplotlib.pyplot as plt
import numpy as np

b = np.arange(2, 5, 0.025)
plt.hist(df['M_inv'], bins = b)
plt.grid(True)
plt.show()
```

```python
b = np.arange(2, 5, 0.025)
plt.xlim(3, 5)
plt.ylim(0, 500)
plt.hist(df['M_inv'], bins = b)
plt.grid(True)
plt.show()
```
