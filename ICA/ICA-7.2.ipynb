## Part 1: Set up a group GitHub repo

You can find more detailed instructions for this process back on ICA 1.2.

The overview is:
1. One group member create a GitHub repo, and add your members as collaborators.
    - Also add the faculty instructor and TA: `hamerski` and `minifold`
2. Every group member create a directory in your JupyterHub, and initiate a local repo there.
3. Link your local repo (JH) to the remote repo (GH).

Once you are in your newly created directory in your terminal, use these commands to initiate a connection to the GitHub repo and set it to be upstream from your local repo:

```
git init
git remote add origin git@github.com:<user>/<repository>.git
git pull --set-upstream origin main
```

You can see PCA 3.1 for more details about what these commands mean, and how you can pull and push successfully from a shared repo.

You can check your setup with these commands:

- `git remote -v`: You should see two lines with the origin SSH URL, one with fetch and one with push. This means the remote GitHub repo has been set up properly as origin.
- `git branch -vv`: You should a line that starts with, "* main" and also contains "\[origin/main\]." This means you set the remote GitHub repo main branch to be upstream from your main branch on your local repo.
